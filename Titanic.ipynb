```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/project/Titanic-Dataset.csv")
data.head()
```

```python

age_series = data["Age"]
fare_series = data["Fare"]
survival_series = data["Survived"]

print(type(age_series))
print(type(data))
```

```python
data.head()
data.info()
data.describe()
```

```python
data.isnull().sum()
data["Age"].fillna(data["Age"].mean(), inplace=True)
data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)
```

```python

data.loc[5]
data.iloc[0:5]
data[["Age", "Fare", "Survived"]]
```

```python
data.rename(columns={"Sex": "Gender"}, inplace=True)
data.pivot_table(values="Survived", index="Gender", columns="Pclass", aggfunc="mean")
```

```python

data[data["Age"] > 30]
data[data["Fare"] > 50]
data[(data["Gender"] == "female") & (data["Survived"] == 1)]
data.sort_values(by="Fare", ascending=False)
data.sort_values(by="Age")
data.sort_values(by="Survived")
```

```python
data.groupby("Gender")["Survived"].mean()
data.groupby("Pclass")["Age"].mean()
data.groupby("Embarked")["Survived"].sum()
```

```python
class_info = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Category": ["Upper", "Middle", "Lower"]
})

merged_data = pd.merge(data, class_info, on="Pclass")
data_part1 = data.iloc[:200]
data_part2 = data.iloc[200:]

pd.concat([data_part1, data_part2])
```

```python
data.groupby("Survived")["Age"].mean()
data["Fare"].max()
data["Fare"].min()
survival_percentage = (data["Survived"].mean()) * 100
print("Survival Percentage:", survival_percentage)
```
